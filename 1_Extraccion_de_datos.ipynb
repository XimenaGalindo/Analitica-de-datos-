```python
df= pd.read_csv("CountryTable.csv", index_col=1)
df
```

```python
#filtro de FILAS VS COLUMNAS
filtro4=df.loc[["Angola","Aruba"], ["life_expectancy"]]
filtro4
```

```python
#filtro de CABECERA
filtro5=df.head(3)
filtro5
```

```python
#filtro de COMPARACION
filtro6=df[df["gnp"] > 828]
filtro6
```

```python
#Filtro Y
filtro7=df[(df["gnp"] > 828) & (df["government_form"] == "Republic")]
filtro7
```

```python
#Filtro O
filtro8=df[(df["gnp"] > 828) | (df["government_form"] == "Republic")]
filtro8
```

```python
#Filtro NOT
filtro9=df[~(df["gnp"] >= 828) & ~(df["government_form"] == "Republic")]
filtro9
```

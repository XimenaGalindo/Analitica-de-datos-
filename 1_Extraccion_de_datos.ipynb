```python
df= pd.read_csv("CountryTable.csv", index_col=1)
df
```

```python
#filtro de FILAS VS COLUMNAS
filtro4=df.loc[["American Samoa","Austria"], ["population"]]
filtro4
```

```python
#filtro de CABECERA
filtro5=df.head(5)
filtro5
```

```python
#filtro de COMPARACION
filtro6=df[df["surface_area"] > 958]
filtro6
```

```python
#Filtro Y
filtro7=df[(df["surface_area"] > 958) & (df["government_form"] == "Republic")]
filtro7
```

```python
#Filtro O
filtro8=df[(df["surface_area"] > 958) | (df["government_form"] == "Republic")]
filtro8
```

```python
#Filtro NOT
filtro9=df[~(df["surface_area"] >= 828) & ~(df["government_form"] == "Republic")]
filtro9
```
